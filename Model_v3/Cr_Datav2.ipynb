```python
import numpy as np
import matplotlib
from mido import Message, MetaMessage
import mido
import sys
import os

def get_pwd(pwdb):
#     pwd = input()
    pwd = ""
    if(not pwd):
        pwd = pwdb
    return pwd

pwd = get_pwd("/home/zedroman/Documents/Sonia_Data/Songs/rap/")
save_pwd=get_pwd("/home/zedroman/Documents/Sonia_Data/Songs/anime_samples2/")
print(pwd)
print(save_pwd)
velocity = [0, 64]
time_range = [0, 150, 300];
tempo = 2000000 // 3
pedal_max = 20
max_len = 2000
use_pedal = False
st = 0.0013354687499999999
```

```python
def is_midi(f):
    if (f[len(f)-5:len(f)] == ".midi"):
        return True
    if (f[len(f)-4:len(f)] == ".mid"):
        return True
    return False
```

```python
music = [f for f in os.listdir(pwd) if is_midi(f)]
music = music
print("\n".join(music))
```

```python
def print_midi(midi_file):
    for track in midi_file.tracks:
        for msg in track:
            print(msg)
        
def find_nearest(n, vec):
    for i in range(len(vec) - 1):
        if (abs(n - vec[i]) < abs(n - vec[i + 1])):
            return vec[i]
    return vec[-1]
```

```python
def convert(pwd, num, savepwd):
    midi_file = mido.MidiFile(pwd)
    concat = mido.MidiFile(type=0)
    tr = mido.MidiTrack()
    concat.tracks.append(tr)
    tr.append(MetaMessage("set_tempo", tempo=tempo, time=0))
    time = 0;
    notes = []
    last_notes = []
    pedal = pedal_max
    now_len = 0
    for msg in midi_file.play():
        if (msg.type != "note_on"):
            continue
#         print(msg)
#         print(int(mido.second2tick(track.time, ticks_per_beat=90, tempo=tempo)))
        now_len += 1
        if (use_pedal):
            pedal -= 1
            if (pedal == 0):
                for i in range(len(last_notes) - 1):
                    tr.append(Message("note_off", note=last_notes[i], velocity=0, time=0, channel=0))
                last_notes = [last_notes[-1]]
                pedal = pedal_max

        if (use_pedal):
            last_notes.append(msg.note)
        tr.append(Message("note_on", note=msg.note, velocity=find_nearest(msg.velocity, velocity), time=find_nearest(msg.time / st, time_range), channel=0))

        if (now_len >= max_len):
            if(use_pedal):
                for i in last_notes:
                    tr.append(Message("note_off", note=i, velocity=0, time=0, channel=0))
            break
#         break
    print(concat.length)
    concat.save(savepwd + str(num) + ".mid")
    num += 1
    return num
```
