```python
import numpy as np
import matplotlib
from mido import Message, MetaMessage
import mido
import sys
import os

def get_pwd(pwdb):
#     pwd = input()
    pwd = ""
    if(not pwd):
        pwd = pwdb
    return pwd

pwd = get_pwd("/home/zedroman/Documents/Sonia_Data/Songs/Naruto/")
save_pwd=get_pwd("/home/zedroman/Documents/Sonia_Data/Songs/anime_samples/")
print(pwd)
print(save_pwd)
velocity = 64
time_range = 240;
tempo = 400000
pedal_max = 10
max_len = 500
```

```python
def is_midi(f):
    if (f[len(f)-5:len(f)] == ".midi"):
        return True
    if (f[len(f)-4:len(f)] == ".mid"):
        return True
    return False
```

```python
music = [f for f in os.listdir(pwd) if is_midi(f)]
print("\n".join(music))
```

```python
def print_midi(midi_file):
    for track in midi_file.tracks:
        for msg in track:
            print(msg)
```

```python
def convert(pwd, num, savepwd):
    midi_file = mido.MidiFile(pwd)
    concat = mido.MidiFile(type=0)
    tr = mido.MidiTrack()
    concat.tracks.append(tr)
    tr.append(MetaMessage("set_tempo", tempo=tempo, time=0))
    time = 0;
    notes = []
    last_notes = []
    pedal = pedal_max
    now_len = 0
    for j, track in enumerate(midi_file.tracks):
        for msg in track:
            if (msg.type != "note_on"):
                continue
            if (time + msg.time > time_range):
                now_len += 1
                pedal -= 1
                if (pedal == 0):
                    for i in range(len(last_notes) - 1):
                        tr.append(Message("note_off", note=last_notes[i], velocity=velocity, time=0, channel=0))
                    last_notes = [last_notes[-1]]
                    pedal = pedal_max
                for i in notes:
                    last_notes.append(i)
                    tr.append(Message("note_on", note=i, velocity=velocity, time=time_range, channel=0))
                    break
#                 tr.append(Message("note_on", note = 0, velocity = 0, time = time_range, channel=0))
#                 tr.append(Message("note_off", note = 0, velocity = 0, time = 0, channel=0))
                for i in notes:
#                     tr.append(Message("note_off", note=i, velocity=velocity, time=time_range, channel=0))
                    break
                notes = []
                time = msg.time - (time_range - time)
#                 while(time > time_range and time < 100):
#                     print(time)
#                     tr.append(Message("note_on", note = 0, velocity = 0, time = time_range, channel=0))
#                     tr.append(Message("note_off", note = 0, velocity = 0, time = 0, channel=0))
#                     time -= time_range
                time -= msg.time
                
            notes.append(msg.note)
            time += msg.time
            if (now_len >= max_len):
                for i in last_notes:
                    tr.append(Message("note_off", note=i, velocity=velocity, time=0, channel=0))
                break
#         break
    print(concat.length)
    concat.save(savepwd + str(num) + ".mid")
    num += 1
    return num
```
